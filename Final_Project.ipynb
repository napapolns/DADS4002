```python
f = open("average data.txt","w")
print('>> ข้อมูลการใช้เวลาโดยเฉลี่ยของคุณ <<')
print('>> ข้อมูลการใช้เวลาโดยเฉลี่ยของคุณ <<',file = f)
##### เช็ครหัสเข้าสู่ระบบถูกต้องหรือไม่
check_id_tf = False
password_tf1 = False
password_tf = ()
#เลขบัตร ปชช
while check_id_tf == False or password_tf1 == False:
    while True:
        try:
            ask_id = input("User ID : ")
            ask_id1 = int(ask_id)
            if ask_id1 > 9999999999999:
                print('กำหนดใหม่\n')
            elif ask_id1 < 1000000000000:
                print('กรุณากำหนดใหม่\n')
            else:
                break
        except : print("กรุณากรอกตัวเลข\n")

#pass
    while True:
        try:
            ask_pass = input("Password : ")
            ask_pass1 = int(ask_pass)
            if ask_pass1 > 9999:
                print('กำหนดใหม่\n')
            elif ask_pass1 <= 0:
                print('กำหนดใหม่\n')
            else:
                break
        except : print("กรุณากรอกตัวเลข\n")

#ตรวงจสอบเลขบัตรและพาส ใน sql
    import mysql.connector

    connection = mysql.connector.connect(
        host="localhost",
        user="root",
        password="2310691201",
        database="project")
        
    mySql_query = "select id_user, password from personal"
    cursor = connection.cursor()
    cursor.execute(mySql_query)
    result = cursor.fetchall()
    
    for data in result:
        if data[0] == ask_id:
            check_id_tf = True
            password_tf = data[1]
            break
    if check_id_tf == True:
        if ask_pass == password_tf:
            password_tf1 = True
            break
        else:
            print('User ID หรือ Password ผิด!!\n')
            password_tf1 = False
            continue
    else:
        print('User ID หรือ Password ผิด!!\n')
        continue


#ดึงข้อมูลการนอนมาโชว์        
mySql_query = "select avg(sleep_h*60 + sleep_m) as avg_sleep from sleep where id_user = %s"
v = (ask_id,)
cursor = connection.cursor()
cursor.execute(mySql_query,v)
result = cursor.fetchall()

result1 = list(result[0])
avg_sleep_h = int((result1[0])//60)
avg_sleep_m = int((result1[0] - (avg_sleep_h*60)))
print('\n>> โดยเฉลี่ยแล้วคุณนอนหลับ', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที')
print('\n>> โดยเฉลี่ยแล้วคุณนอนหลับ', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที',file = f )

if avg_sleep_h < 7 :
    print('-->> คุณนอน', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที คุณนอนน้อยเกินไป คุณควรนอนให้มากขึ้น')
    print('-->> คุณนอน', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที คุณนอนน้อยเกินไป คุณควรนอนให้มากขึ้น',file = f )
elif avg_sleep_h > 9 :
    print('-->> คุณนอน', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที คุณนอนมากเกินไป')
    print('-->> คุณนอน', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที คุณนอนมากเกินไป',file = f  )
elif avg_sleep_h == 9 and avg_sleep_m > 0 :
    print('-->> คุณนอน', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที คุณนอนมากเกินไป')
    print('-->> คุณนอน', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที คุณนอนมากเกินไป',file = f  )
elif (avg_sleep_h >= 7 and avg_sleep_m >= 0) or (avg_sleep_h < 9 and avg_sleep_m >= 0):
    print('-->> คุณนอน', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที คุณนอนเหมาะสม')
    print('-->> คุณนอน', avg_sleep_h, 'ชั่วโมง', avg_sleep_m, 'นาที คุณนอนเหมาะสม',file = f  )

#ดึงข้อมูลการทำงานมาโชว์        
mySql_query = "select avg(work_h*60) as avg_working from working where id_user = %s"
v = (ask_id,)
cursor = connection.cursor()
cursor.execute(mySql_query,v)
result = cursor.fetchall()

result1 = list(result[0])
avg_work_h = int((result1[0])//60)
print('\n>> โดยเฉลี่ยแล้วคุณทำงาน', avg_work_h, 'ชั่วโมง')
print('\n>> โดยเฉลี่ยแล้วคุณทำงาน', avg_work_h, 'ชั่วโมง',file = f )

if avg_work_h < 7 :
    print('-->> คุณทำงาน', avg_work_h, 'ชั่วโมง คุณทำงานเสร็จเร็ว')
    print('-->> คุณทำงาน', avg_work_h, 'ชั่วโมง คุณทำงานเสร็จเร็ว',file = f  )
elif avg_work_h > 8 :
    print('-->> คุณทำงาน', avg_work_h, 'ชั่วโมง คุณทำงานมากเกินไป')
    print('-->> คุณทำงาน', avg_work_h, 'ชั่วโมง คุณทำงานมากเกินไป',file = f  )
else:
    print('-->> คุณทำงาน', avg_work_h, 'ชั่วโมง คุณทำงานเหมาะสม')
    print('-->> คุณทำงาน', avg_work_h, 'ชั่วโมง คุณทำงานเหมาะสม',file = f  )

#ดึงข้อมูลการออกกำลังกายมาโชว์        
mySql_query = "select avg(exercise_h*60 + exercise_m) as avg_exercise from exercise where id_user = %s"
v = (ask_id,)
cursor = connection.cursor()
cursor.execute(mySql_query,v)
result = cursor.fetchall()

result1 = list(result[0])
avg_exercise_h = int((result1[0])//60)
avg_exercise_m = int((result1[0] - (avg_exercise_h*60)))
print('\n>> โดยเฉลี่ยแล้วคุณออกกำลังกาย', avg_exercise_h, 'ชั่วโมง', avg_exercise_m, 'นาที')
print('\n>> โดยเฉลี่ยแล้วคุณออกกำลังกาย', avg_exercise_h, 'ชั่วโมง', avg_exercise_m, 'นาที',file = f )

if avg_exercise_h >= 1 or avg_exercise_m >= 15 :
    print('-->> คุณออกกำลังกาย', avg_exercise_h, 'ชั่วโมง', avg_exercise_m, 'นาที ดีมาก ทำต่อไปนะ!!')
    print('-->> คุณออกกำลังกาย', avg_exercise_h, 'ชั่วโมง', avg_exercise_m, 'นาที ดีมาก ทำต่อไปนะ!!',file = f  )
elif avg_exercise_m == 0 :
    print('-->> คุณออกกำลังกาย', avg_exercise_h, 'ชั่วโมง', avg_exercise_m, 'นาที ไปออกกำลังกายเดี๋ยวนี้!!')
    print('-->> คุณออกกำลังกาย', avg_exercise_h, 'ชั่วโมง', avg_exercise_m, 'นาที ไปออกกำลังกายเดี๋ยวนี้!!',file = f  )
else:
    print('-->> คุณออกกำลังกาย', avg_exercise_h, 'ชั่วโมง', avg_exercise_m, 'นาที คุณออกกำลังกายน้อยเกินไปนะ ออกเพิ่มหน่อยดีไหม')
    print('-->> คุณออกกำลังกาย', avg_exercise_h, 'ชั่วโมง', avg_exercise_m, 'นาที คุณออกกำลังกายน้อยเกินไปนะ ออกเพิ่มหน่อยดีไหม',file = f  )
f.close()
```
